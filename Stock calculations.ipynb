```python
stock_list=["AAPL.csv",'GOOG.csv','SAF.PA.csv','AIR.DE.csv','LMT.csv','BA.csv','UTC1.DE.csv','BBD-B.TO.csv','GE.csv','RR.L.csv']
```

```python
close_df = pd.DataFrame()
for stock in stock_list:
#     close = pd.read_csv(stock, usecols=['Close'])
    close_df[stock]=pd.read_csv(stock, usecols=['Close'])
#     print(close.head())
close_df.dropna(inplace=True)
close_df
```

```python
close_df_2=pd.DataFrame()
for col in close_df:
    close_df_2[col]=close_df[col]/close_df[col].mean()
close_df_2
```

```python
np.cov(close_df_2.transpose())
```

```python
close_df.mean()
```

```python
# final_df=pd.DataFrame(columns=['Return','var_1','var_2','var_3','var_4','var_5','var_6','var_7','var_8','var_9','var_10'])
final_df=pd.DataFrame(columns=['Return'])
```

```python
final_df['Return']=close_df.mean()/1000
```

```python
final_df.reset_index(drop=True,inplace=True)
final_df
```

```python
for i in range(10):
#     print(i)
    final_df['var_'+str(i+1)]=np.cov(close_df_2.transpose())[i]
```
