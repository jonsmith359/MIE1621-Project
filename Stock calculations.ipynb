```python
stock_list=["AAPL.csv",'GOOG.csv','SAF.PA.csv','AIR.DE.csv','LMT.csv','BA.csv','UTC1.DE.csv','BBD-B.TO.csv','GE.csv','RR.L.csv']
```

```python
# close_df = pd.DataFrame()
close_df = pd.DataFrame(columns=stock_list)#, usecols=stock_list)
print(close_df)
for stock in stock_list:
#     close = pd.read_csv(stock, usecols=['Close'])
    close_df[stock]=pd.read_csv(stock, usecols=['Close'])
#     print(close.head())
close_df.dropna(inplace=True)
```

```python
final_df['Return']=close_df.mean()
```
